```python
# prompt: write a python function to generate a random date between a start_date variable and an end_date variable.

import random
from datetime import datetime, timedelta

def random_date(start_date, end_date):
  """
  Generate a random date between two given dates.

  Args:
    start_date (datetime): The start date.
    end_date (datetime): The end date.

  Returns:
    datetime: A random date between the start and end dates.
  """

  # Convert the start and end dates to seconds since the epoch.
  start_date_seconds = start_date.timestamp()
  end_date_seconds = end_date.timestamp()

  # Calculate the difference between the start and end dates in seconds.
  date_range_seconds = end_date_seconds - start_date_seconds

  # Generate a random number of seconds within the date range.
  random_seconds = random.random() * date_range_seconds

  # Add the random seconds to the start date to get a random date.
  random_date = start_date + timedelta(seconds=random_seconds)

  return random_date
```

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Setting a seed for reproducibility
random.seed(42)
np.random.seed(42)

# Function to generate random dates within a specified range
def random_dates(start_date, end_date, n):
    return [random_date(start_date, end_date) for _ in range(n)]

# Example usage
# Constants
n = 100  # number of entries
start_date = datetime.now() - timedelta(days=365)
end_date = datetime.now()

# Data generation
data = {
    "Transaction ID": range(1001, 1001+n),
    "Card Type": np.random.choice(['Visa', 'MasterCard', 'American Express'], n),
    "Date": random_dates(start_date, end_date, n),
    "Time": [datetime.now().replace(microsecond=0, second=0, minute=0) + timedelta(minutes=15*i) for i in range(n)],
    "Amount": np.random.uniform(0.5, 500.0, n).round(2),
    "Vendor": np.random.choice(['Amazon', 'Walmart', 'Apple', 'Nike', 'Costco', 'Starbucks', 'McDonalds'], n),
    "Location": np.random.choice(['New York, USA', 'London, UK', 'Paris, FR', 'Berlin, DE', 'Tokyo, JP'], n),
    "Category": np.random.choice(['Groceries', 'Electronics', 'Clothing', 'Dining', 'Entertainment'], n),
    "Status": np.random.choice(['Approved', 'Declined'], n, p=[0.95, 0.05]),
    "Fraud Flag": np.random.choice(['Yes', 'No'], n, p=[0.05, 0.95])
}

# Create DataFrame
df = pd.DataFrame(data)
df['Time'] = df['Date'].dt.strftime('%Y-%m-%d') + ' ' + df['Time'].dt.strftime('%H:%M:%S')  # Concatenate date and time for full timestamp
df.drop(columns='Date', inplace=True)  # Drop the separate date column

# Save to CSV
df.to_csv('Credit_Card_Transactions.csv', index=False)
df
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# Set seed for reproducibility
np.random.seed(42)

# Helper functions
def random_date(start, end):
    """Generate a random datetime between `start` and `end`."""
    return start + timedelta(
        seconds=random.randint(0, int((end - start).total_seconds())))

# Define data size
n = 1000

# Generate data
data = {
    "Transaction ID": [f"TX{1000+i}" for i in range(n)],
    "Timestamp": [random_date(datetime(2023, 1, 1), datetime(2023, 12, 31)) for _ in range(n)],
    "Amount": [np.abs(np.random.normal(50, 25)) if i % 50 != 0 else np.abs(np.random.normal(200, 100)) for i in range(n)],  # anomalous amounts
    "Merchant": [random.choice(["Amazon", "Walmart", "Target", "Starbucks", "Apple Store"]) for _ in range(n)],
    "Category": [random.choice(["Electronics", "Groceries", "Dining", "Clothing", "Cafe"]) for _ in range(n)],
    "User ID": [f"U{100+i%100}" for i in range(n)],  # 100 users
    "Card Type": [random.choice(["Visa", "MasterCard", "Amex", "Discover"]) for _ in range(n)],
    "Location": [random.choice(["New York", "California", "Texas", "Florida", "Illinois"]) if i % 100 != 0 else "Nevada" for i in range(n)],  # anomalous locations
    "Fraud Flag": [0 if i % 50 != 0 else 1 for i in range(n)]  # 2% fraud rate
}

# Create DataFrame
df = pd.DataFrame(data)

# Sort by Timestamp
df.sort_values(by="Timestamp", inplace=True)

# Reset index for clarity
df.reset_index(drop=True, inplace=True)

# Show a sample of the data and save to CSV
df.head(), df.to_csv("Credit_Card_Transactions_with_anomalies.csv", index=False)
```

## Anomalies Detection Unsupervised

```python
def enhance_timestamp(df, timestamp_column):
    """
    Enhances the input DataFrame by adding new columns derived from a timestamp column.

    Parameters:
        df (pd.DataFrame): The input DataFrame.
        timestamp_column (str): The name of the timestamp column to be enhanced.

    Returns:
        pd.DataFrame: A new DataFrame containing the original data along with the new columns.
    """
    # Ensure the specified column is in datetime format
    df[timestamp_column] = pd.to_datetime(df[timestamp_column])

    # Create new columns based on the timestamp column
    df['month'] = df[timestamp_column].dt.month
    # df['minute'] = df[timestamp_column].dt.minute
    df['hour'] = df[timestamp_column].dt.hour
    df['day_of_week'] = df[timestamp_column].dt.weekday

    return df

def convert_to_label_encoding(df, categorical_columns):
    """
    Converts specified categorical columns in the DataFrame to label encoding.

    Parameters:
        df (pd.DataFrame): The input DataFrame.
        categorical_columns (list): List of column names to be label encoded.

    Returns:
        pd.DataFrame: A new DataFrame with the specified columns label encoded.
    """
    # Make a copy of the DataFrame to avoid modifying the original
    new_df = df.copy()

    # Iterate over the specified columns and apply label encoding
    for column in categorical_columns:
        new_df[column + '_Label'] = new_df[column].astype('category').cat.codes

    return new_df



```

```python
from pyod.models.iforest import IForest
import numpy as np

import pandas as pd
def get_anomaly_records():
    df = pd.read_csv('Credit_Card_Transactions_with_anomalies2.csv')
    df.to_csv('result_Credit_Card_Transactions_with_anomalies2.csv', index=False)
    new_df = df.copy()


    handle_timestamp_df = enhance_timestamp(new_df, 'Timestamp')
    handle_categorical_df = convert_to_label_encoding(handle_timestamp_df, ['Merchant', 'Category','User ID', 'Card Type', 'Location'])
    final_df = handle_categorical_df.drop(['Transaction ID', 'Timestamp', 'Merchant', 'Category','User ID', 'Card Type', 'Location'], axis=1)

    outliers_fraction = 0.02
    random_state = np.random.RandomState(42)
    model_iso = IForest(contamination=outliers_fraction,random_state=random_state)
    model_iso.fit(final_df)

    # Predict anomalies (-1 for outliers, 1 for inliers)
    predictions = model_iso.predict(final_df)
# Add the predictions to the original DataFrame (for interpretability)
    df['predicted_anomaly'] = predictions

    # Filter to show only the anomalies
    anomalies = df[df['predicted_anomaly'] == 1]
    anomalies['predicted_anomaly'] = [True if i == 1 else False for i in anomalies['predicted_anomaly']]
    print(anomalies)

    tranform_data_anomalies = final_df.copy()
    tranform_data_anomalies['predicted_anomaly'] = predictions
    tranform_data_anomalies_only = tranform_data_anomalies[tranform_data_anomalies['predicted_anomaly'] == 1]
    tranform_data_anomalies_only['predicted_anomaly'] = [True if i == 1 else False for i in tranform_data_anomalies_only['predicted_anomaly']]
    tranform_data_anomalies_only = tranform_data_anomalies_only.reset_index()
    # print(f"Total anomalies detected: {anomalies.shape}")
    # df.to_csv('Credit_Card_Transactions_with_anomalies2.csv', index=False)
    return anomalies, final_df, df, model_iso, tranform_data_anomalies_only
```
